##  Think Like a Researcher

Suppose two retrieval systems return different results.

Which one is better?

Instead of guessing,

we calculate objective evaluation metrics.

These metrics tell us how effectively a retriever finds relevant documents.

```python
relevant_documents = {
    "Doc A",
    "Doc C",
    "Doc E"
}

retrieved_documents = [
    "Doc A",
    "Doc B",
    "Doc D",
    "Doc C",
    "Doc F"
]
```

## Precision@K

Precision measures:

Out of the retrieved documents,

how many are relevant?

Formula

Precision@K = Relevant Retrieved 
/
Retrieved Documents


or

Precision@K  =  Number of Relevant Documents in Top K​​  / K

```python
k = 5

retrieved = retrieved_documents[:k]

correct = len(
    set(retrieved).intersection(
        relevant_documents
    )
)

precision = correct / k

print(precision)

# 2/5
```

## Recall@K

Recall measures:

Out of all relevant documents,

how many did we retrieve?

Formula

Recall@K =
Relevant Retrieved
/
Total Relevant Documents

or 



Recall@K=
Number of Relevant Documents Retrieved in Top K
/
Total Number of Relevant Documents
	​

	​


```python
recall = correct / len(relevant_documents)

print(recall)
```

## Hit Rate

Hit Rate answers one simple question.

Did we retrieve

at least one

relevant document?

Yes

↓

1

No

↓

0

```python
hit_rate = int(correct > 0)

print(hit_rate)
```

## Mean Reciprocal Rank (MRR)

MRR rewards systems

that return a relevant document early.

Formula

MRR =
1
/
Rank of First Relevant Document

```python
for i,doc in enumerate(retrieved_documents,1):

    if doc in relevant_documents:

        mrr = 1/i

        break

print(mrr)
```

## NDCG

NDCG

(Normalized Discounted Cumulative Gain)

considers

- relevance

and

- ranking position.

Relevant documents appearing earlier receive higher scores.

```python
from sklearn.metrics import ndcg_score

true = [[1,0,1,1,0]]

pred = [[0.95,0.80,0.60,0.55,0.10]]

score = ndcg_score(
    true,
    pred
)

print(score)
```

## Comparison

| Metric | Measures |
|---------|----------|
| Precision | Retrieved relevance |
| Recall | Coverage of relevant documents |
| Hit Rate | At least one correct result |
| MRR | Position of first relevant result |
| NDCG | Overall ranking quality |

## Which Metric Should You Use?

Need accurate Top-K results?

↓

Precision@K

Need to retrieve all relevant documents?

↓

Recall@K

Need users to quickly see one useful result?

↓

MRR

Need overall ranking quality?

↓

NDCG
